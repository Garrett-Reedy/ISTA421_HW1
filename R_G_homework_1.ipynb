```python
#QUESTION 6

class w: #weight stats
    weight = np.empty(100) #creates empty array
    mean = 81
    sd = 0.26
    units = "kg"

class h: #height stats
    height = np.empty(100) #creates empty array
    mean = 1.76
    sd = 0.22
    units = "m"

class a: #age stats
    age = np.empty(100) #creates empty array
    mean = 25
    sd = 5
    units = "years"

gender = np.empty(100, str)
genderOpts = ["A", "B", "C"]

for i in range(100):
    w.weight[i] = np.random.normal(w.mean, w.sd)
    h.height[i] = np.random.normal(h.mean, h.sd)
    a.age[i] = np.random.normal(a.mean, a.sd)
    gender[i] = genderOpts[np.random.randint(0,len(genderOpts))] #randomly samples from list ["A", "B", "C"]

print(w.weight)
print(h.height)
print(a.age)
print(gender)
    
```

```python
#QUESTION 7

BMI = np.full(100, np.nan) #generates 100 element array of NaN

Q7 = pd.DataFrame({ #Organizes the data by column in this dataframe
    "Weight": w.weight, 
    "Height": h.height, 
    "Age": a.age, 
    "Gender": gender,
    "BMI": BMI
    }
)

print(Q7)
```

```python
#QUESTION 8

AgeGroup = np.empty(100, 'U8') #creates array of 8 character strings

for i in range(100): 
    temp = Q7.at[i, "Age"] #references the ages 1 by 1

    if temp < 20:
        AgeGroup[i] = "Under 20"
    elif temp > 30:
        AgeGroup[i] = "Over 30"
    else:
        AgeGroup[i] = "20 to 30"

Q8 = Q7.copy() #copies Q7
Q8["Age group"] = AgeGroup #Adds column titled "Age group" to Q8

print(Q8)
```

```python
#QUESTION 9

Q9 = Q8.copy() #copies Q8

for i in range(100):
    Q9.at[i, "BMI"] = Q9.at[i, "Weight"] / (pow(Q9.at[i, "Height"], 2)) #calculates BMI for each index

print(Q9)
```

```python
#QUESTION 10

def function_Q10(f_weight, f_height):
    f_BMI = f_weight / pow(f_height, 2) #height / weight^2
    return f_BMI #returns calculated BMI

print(function_Q10(w.weight, h.height)) #proof that it works



```

```python
#QUESTION 11

def function_Q11(Q9, gender):
    sum = 0 #BMI sum for gender
    num = 0 #total count of that gender

    for i in range(100):
        if Q9.at[i, "Gender"] == gender: 
            sum += Q9.at[i,"BMI"] #adds to sum
            num += 1 #increments count
    
    return sum / num #Average BMI

print(function_Q11(Q9, "A")) #proof that it works

```
